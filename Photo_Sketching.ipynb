```python
import cv2 as cv
from PIL import Image
```

```python
img = cv.imread('faces.jpg')

#cv.pencilSketch() will return a tuple (2 elements) gray sketch & color sketch
#shade_factor( range 0 to 0.1 )
gray, color = cv.pencilSketch(img, sigma_s = 60, sigma_r = 0.07, shade_factor = 0.05)

#displaying output of color skrtch using cv.imshow, must execute cv.waitKey & cv.destroyAllWindows otherwise kernel will crash
cv.imshow('color', color)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#displaying grayscale result using PIL's Image module
Image.fromarray(gray)
```

```python
#displaying color result using PIL's Image module. (Warning: object  created by cv (i.e: ndarray) should not displayed by PIL's Image Module, see the result)
Image.fromarray(color)
```
